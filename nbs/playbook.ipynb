## Visualizing Manifolds

PCA visualization of learned representations from different training objectives.

#### Supervised Learning

```python
import pathlib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load embeddings
base = pathlib.Path('../representation/mnist/supervised/step_50000')
x = np.load(base / 'embeddings.npy')
y = np.load(base / 'labels.npy')
```

```python
pca = PCA(n_components=2)
x2 = pca.fit_transform(x)

plt.figure(figsize=(6, 6))
plt.scatter(x2[:, 0], x2[:, 1], c=y, cmap='tab10', s=10, alpha=0.6)
plt.colorbar(label='Digit Class')
plt.title('PCA of Supervised MNIST embeddings')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

## Autoencoder

Learns to reconstruct input images.

```python
from torchvision import datasets

# Load embeddings
base = pathlib.Path('../representation/mnist/autoencoder/step_50000')
x = np.load(base / 'embeddings.npy')

# Load actual labels (labels.npy contains images for autoencoder)
mnist_test = datasets.MNIST(root='../data', train=False, download=True)
rng = np.random.default_rng(42)
indices = rng.choice(len(mnist_test), 10000, replace=False)
y = np.array([mnist_test[i][1] for i in indices])
```

```python
pca = PCA(n_components=2)
x2 = pca.fit_transform(x)

plt.figure(figsize=(6, 6))
plt.scatter(x2[:, 0], x2[:, 1], c=y, cmap='tab10', s=10, alpha=0.6)
plt.colorbar(label='Digit Class')
plt.title('PCA of AutoEncoder MNIST embeddings')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
